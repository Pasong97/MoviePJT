## (1) 형태소 분석기로만(예시일 뿐 사용하지 않음)



```python
# 전처리 함수 정의
def preprocessing(stemmer, sentences):
    review_lst = []
    print('|Start!', end=' ')
    for i, sentence in enumerate(sentences):
        sentence = re.sub('[^0-9가-힣ㄱ-ㅎㅏ-ㅣ]', '', sentence)        # 특수문자 제거
        words = stemmer.morphs(sentence)                               # 토크나이징
        words = [w for w in words if w not in stopwords]               # 불용어 제거 
        # words = ' '.join(words)                                      # 다시 병합

        review_lst.append(words)

        if (i + 1) % (len(sentences) // 10) == 0:
            print(f'> {i+1}', end=' ')     
    print('End!|')

    return review_lst
```

```python
def SearchNgram_okt(n, elements, knu_dict):
    gram_df = knu_dict[knu_dict['pos_len']==n]

    if n == 1:
        elements = [[e] for e in elements]
    else:
        elements = list(zip(*[review[i:] for i in range(n)]))
        elements = [list(e) for e in elements]
    # print(elements)
    
    search_lst = []
    search_cnt = 0
    for i, element in enumerate(elements):
        for j, word in enumerate(gram_df['okt_word']):
            if f'{element}' == word:
                search_cnt += 1
                pol = gram_df.iloc[j, 1]
                search_lst.append([i, element, pol])
    
    # 메세지 출력
    if search_cnt != 0:
        print(f'> {n} gram O({search_cnt})', end=' ')
    else:
        print(f'> {n} gram X', end=' ')   

    return search_lst
```

```python
details = []
polarity = []
for i, review in enumerate(okt_reviews):
    print(f'> {i+1}번째 review 탐색', end='')
    
    # ngram search
    all_search_lst = []
    for n in range(2,0,-1):
        search_lst = SearchNgram_okt(n, review, knu_dict)
        if n == 2 and search_lst != []:
            idx = [lst[0] for lst in search_lst]
        elif n == 1 :
            for x in idx:
                search_lst = [lst for lst in search_lst if lst[0] not in [x,x+1]]
        all_search_lst.extend(search_lst)
        
    pol_sum = sum([lst[-1] for lst in search_lst])
    details.append(all_search_lst)
    polarity.append(pol_sum)

    print(f'> 탐색 종료!')
    # print(details[i])

data['okt_details'] = details
data['okt_polarity'] = polarity

data
```

## sklearn 사용

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation 
```

```python
count_vectorizer = CountVectorizer(max_df=0.1, max_features=1000, min_df=2, ngram_range=(1, 2))

feat_vect = count_vectorizer.fit_transform(texts)
print('CountVectorizer Shape', feat_vect.shape)
```

```python
lda = LatentDirichletAllocation(n_components=3)
lda.fit(feat_vect)
```

```python
feature_names = count_vectorizer.get_feature_names()
feature_names[:10]
```

```python
for topic_index, topic in enumerate(lda.components_):
    print('Topic #', topic_index)

    topic_word_indexs = topic.argsort()[::-1]
    top_indexs = topic_word_indexs[:20]

    feature_concat = ' '.join([feature_names[i] for i in top_indexs])
    print(feature_concat)
```

```python
doc_topic = lda.transform(feat_vect)

doc_per_topic_list = []
for n in range(doc_topic.shape[0]):
    topic_most_pr = doc_topic[n].argmax()
    topic_pr = doc_topic[n].max()
    doc_per_topic_list.append([n, topic_most_pr, topic_pr])
    
doc_topic_df = pd.DataFrame(doc_per_topic_list, columns=['Doc_Num', 'Topic', 'Percentage'])

top_pr_topics = doc_topic_df.join(data)
top_pr_topics
```

## gensim 사용

```python
# bigram 생성에 필요한 library
from gensim.models import Phrases
from gensim.models.phrases import Phraser

# vectorize & lda에 필요한 library
from gensim import corpora
from gensim.models.ldamodel import LdaModel
```

```python
# bigram
# Build the bigram models
bigram = Phrases(texts, min_count=2, threshold=10)
bigram_mod = Phraser(bigram)
 
# See example
print(bigram_mod[texts[0]])
```

```python
# tokenized_list의 모든 문서에 대해 bigram을 생성해줌
words_bigram = [bigram_mod[doc] for doc in texts]
```

```python
dictionary = corpora.Dictionary(words_bigram)  # 각 단어에 번호를 할당해줌
    # bigram 포함하는 과정을 생략하고 싶으면, 그냥 바로 여기에 tokenized_list를 넣어주면 됨

dictionary.filter_extremes(no_below=2, no_above=0.05)   
    # 2개의 문서 미만으로 등장하는 단어는 제외, 전체의 5% 이상으로 자주 등장하는 단어는 제외

corpus = [dictionary.doc2bow(text) for text in words_bigram]
print(corpus[0]) # 첫번째 corpus를 테스트로 출력: 몇 번째 단어가 몇 번 나왔는지 저장되어 있음
```

```python
ldamodel = LdaModel(corpus, num_topics=6, id2word=dictionary, passes=20, iterations=500)  # 토픽 수: 6

ldamodel.print_topics(num_words=7)  # num_words=10이 default
```

```python
doc_per_topic_list = []

for n in range(len(corpus)):
    doc_topic = ldamodel[corpus[n]]
    doc_topic = sorted(doc_topic, key=lambda x: (x[1]), reverse=True)
    topic_most_pr = doc_topic[0][0]
    topic_pr = doc_topic[0][1]
    doc_per_topic_list.append([n, topic_most_pr, topic_pr])
    
doc_topic_df = pd.DataFrame(doc_per_topic_list, columns=['Doc_Num', 'Topic', 'Percentage'])

# 실제 review 내용과 join
doc_topic_df = doc_topic_df.join(data)  

doc_topic_df.head()
```
