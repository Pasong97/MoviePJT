## (1) 리뷰에서 명사 단어 추출

### 1) Okt 명사 추출하기

```python
# 전처리 함수 정의
def ExtractNouns(stemmer, sentences):
    nouns_lst = []
    print('|Start!', end=' ')
    for i, sentence in enumerate(sentences):
        nouns = stemmer.nouns(sentence) 
        nouns_lst.extend(nouns)

        if (i + 1) % (len(sentences) // 10) == 0:
            print(f'> {i+1}', end=' ')     
    print('End!|')

    return set(nouns_lst)
```

```python
okt_nouns = ExtractNouns(okt, data['리뷰'])
print(okt_nouns)
```

## (2) soynlp 학습하기

```python
data['리뷰']
```

```python
from soynlp.utils import DoublespaceLineCorpus

corpus = DoublespaceLineCorpus('./data/reviews.txt', iter_sent=True)
print('number of sents = {}'.format(len(corpus)))
```

```python
import urllib.request
from soynlp.word import WordExtractor
from soynlp.noun import LRNounExtractor_v2
```

```python
# 명사 추출(오래 걸림)
noun_extractor = LRNounExtractor_v2()
nouns = noun_extractor.train_extract(corpus)
```

```python
print('number of nouns = {}'.format(len(nouns)))
```

## (3) 명사 단어 추출

```python
selected_nouns = [(noun, score) for noun, score in nouns.items() if len(noun) > 1]
selected_nouns = sorted(selected_nouns, key=lambda x:-x[1].frequency)

sel_nouns_lst = []
for noun, score in selected_nouns:
    sel_nouns = [noun, score[0], score[1]]
    sel_nouns_lst.append(sel_nouns)

sel_nouns_df = pd.DataFrame(sel_nouns_lst, columns=['noun', 'frequency', 'score'])
sel_nouns_df['noun_len'] = sel_nouns_df['noun'].apply(len)
sel_nouns_df[sel_nouns_df['noun_len'] > 1].sort_values(by='frequency', ascending=False)
sel_nouns_df['nouns_v2'] = sel_nouns_df['noun'].apply(lambda x: re.sub('[^가-힣]','',x))
sel_nouns_df = sel_nouns_df[['nouns_v2','noun', 'frequency', 'score']]
sel_nouns_df
```

```python
sel_nouns = sel_nouns_df['nouns_v2']
```

```python
sel_nouns_df.to_csv('./data/sel_nouns_df.txt', 
                     index=False, header=False)
```
